```python
# Load YouTube dataset
youtube_data = pd.read_csv("USvideos.csv")

# Load Airbnb dataset
airbnb_data = pd.read_csv("AB_NYC_2019.csv")
```

```python
# Check for missing values
youtube_data.isnull().sum()
airbnb_data.isnull().sum()
```

```python
# Check for duplicate records
youtube_data.duplicated().sum()
airbnb_data.duplicated().sum()
```

```python
# Impute missing values in YouTube dataset
# Impute with mean for numerical columns
numerical_cols = youtube_data.select_dtypes(include=np.number).columns
youtube_data[numerical_cols] = youtube_data[numerical_cols].fillna(youtube_data[numerical_cols].mean())

print(youtube_data[numerical_cols])
# Impute with mode for categorical columns
categorical_cols = youtube_data.select_dtypes(include='object').columns
youtube_data[categorical_cols] = youtube_data[categorical_cols].fillna(youtube_data[categorical_cols].mode().iloc[0])

# Impute missing values in Airbnb dataset
# Impute with mean for numerical columns
numerical_cols = airbnb_data.select_dtypes(include=np.number).columns
airbnb_data[numerical_cols] = airbnb_data[numerical_cols].fillna(airbnb_data[numerical_cols].mean())

# Impute with mode for categorical columns
categorical_cols = airbnb_data.select_dtypes(include='object').columns
airbnb_data[categorical_cols] = airbnb_data[categorical_cols].fillna(airbnb_data[categorical_cols].mode().iloc[0])
```

```python
# Remove duplicates from YouTube dataset
youtube_data.drop_duplicates(inplace=True)
print("Duplicates removed from Youtube dataset")
# Remove duplicates from Airbnb dataset
airbnb_data.drop_duplicates(inplace=True)
print("Duplicates removed from Airbnb dataset")
```

```python
# Convert date columns to datetime format
youtube_data['trending_date'] = pd.to_datetime(youtube_data['trending_date'], format='%y.%d.%m')
youtube_data['publish_time'] = pd.to_datetime(youtube_data['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
print("Columns converted")

# Rename columns for consistency
airbnb_data.rename(columns={'neighbourhood_group': 'neighborhood_grp', 'neighbourhood': 'neighborhood'}, inplace=True)
print("Columns renamed")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Detect outliers using box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=youtube_data[['views', 'likes', 'dislikes', 'comment_count']])
plt.title('YouTube Dataset - Boxplot of Views, Likes, Dislikes, and Comment Count')

plt.subplot(1, 2, 2)
sns.boxplot(x='room_type', y='price', data=airbnb_data)
plt.title('Airbnb Dataset - Boxplot of Price by Room Type')

plt.tight_layout()
plt.show()
```

```python
# Handle outliers using z-score method or IQR method
# For example, remove outliers using z-score method
from scipy import stats

# Define a threshold for outlier detection (typically z-score > 3 or z-score < -3)
z_threshold = 3

# Calculate z-scores for each numerical column
z_scores_youtube = np.abs(stats.zscore(youtube_data[['views', 'likes', 'dislikes', 'comment_count']]))
print("Youtube z-scores:\n")
print(z_scores_youtube.head(5),"\n")
z_scores_airbnb = np.abs(stats.zscore(airbnb_data[['price']]))
print("Airbnb z-scores:\n")
print(z_scores_airbnb.head(5))
```

```python
# Identify outliers
outliers_youtube = np.where(z_scores_youtube > z_threshold)

outliers_airbnb = np.where(z_scores_airbnb > z_threshold)
```

```python
# Remove outliers
youtube_data_cleaned = youtube_data[(z_scores_youtube < z_threshold).all(axis=1)]
airbnb_data_cleaned = airbnb_data[(z_scores_airbnb < z_threshold)]
```

```python
# Save cleaned datasets to new CSV files
youtube_data.to_csv("cleaned_youtube_data.csv", index=False)
airbnb_data.to_csv("cleaned_airbnb_data.csv", index=False)
```
